# Solución: Subir archivos correctamente y crear Pull Request

Este notebook muestra el flujo correcto para asegurarte de que todos tus archivos se suben al repositorio y cómo crear un Pull Request (PR) en GitHub.

## 1. Verificar archivos no subidos

Usa el siguiente comando en la terminal integrada de VS Code para ver qué archivos no han sido subidos al repositorio:

```bash
git status
```

## 2. Agregar archivos al área de staging con git add .

Ejecuta este comando para agregar todos los archivos modificados y nuevos al área de staging:

```bash
git add .
```

## 3. Realizar commit de los cambios

Haz un commit de los archivos agregados usando un mensaje descriptivo:

```bash
git commit -m "Mensaje descriptivo del commit"
```

## 4. Hacer push de los cambios al repositorio remoto

Envía los commits al repositorio remoto con el siguiente comando (reemplaza `nombre-de-tu-rama` por el nombre real de tu rama):

```bash
git push origin nombre-de-tu-rama
```

## 5. Crear un Pull Request (PR) en GitHub

1. Abre GitHub y navega a tu repositorio.
2. Haz clic en "Compare & pull request" o "New pull request".
3. Selecciona tu rama como origen y `main` como destino.
4. Escribe un título y descripción claros para tu PR.
5. Haz clic en "Create pull request" para enviarlo a revisión.
